```python
import matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

**Wykres liniowy**

```python
_ = plt.plot([0,1], [0, 4])
```

```python
plt.plot([0,1], [0, 4], label='line', color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Wykres')
_ = plt.legend()
```

```python
plt.plot([0,1, 3], [0, 4, 2], label='line', color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Wykres')
_ = plt.legend()
```

```python
x = np.arange(-3, 3, step=0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

**Tworzenie kilku wykresów**

```python
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 3, 2], [3, 4, 5], 'r--')
```

```python
fig = plt.figure(figsize=(12, 8))
plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 3, 2], [3, 4, 5], 'r--')
plt.subplot(133)
plt.plot([0, 1, 1.5], [3, 2, 5], 'ro')
```

**Wykresy słupkowe**

```python
_ = plt.bar(x=[0, 1, 2, 3], height=[3, 5, 2, 4.5])
```

```python
_ = plt.barh(y=['a', 'b', 'c', 'd'], width=[3, 5, 2, 4.5])
```

**Wykresy punktowe**

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2)
```

```python
x3 = np.random.randn(300) * 50
plt.scatter(x=x1, y=x2, s=x3)
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```
