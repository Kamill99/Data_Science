```python
import numpy as np
import pandas as pd
```

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('UBER')
df.head()
```

**Zapis danych**

```python
df.to_csv('uber.csv')
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2022)]
df_nov
```

```python
df_nov.to_csv('uber_nov.csv')
```

**Odczyt danych**

```python
new_df = pd.read_csv('uber_nov.csv')
```
